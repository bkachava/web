### Summary
* Make the timestamps transformations, in the WeatherPy dataset, to visualize the dates in a proper format for the user.

```python
# Dependencies and Setup
import pandas as pd
import time
import numpy as np
from datetime import datetime

# File (CSV)
data_file = "Resources/cities.csv"
output_data_file = "Resources/cities_dt.csv"
```

```python
# Use Pandas to read data
data_file_pd = pd.read_csv(data_file, encoding="UTF-8")   
data_file_pd.head()
# Display a statistical overview of the DataFrame
data_file_pd.describe()
```

```python
# Display the data frame info
data_file_pd.info()
```

```python
# Make a list of the timestamps
timestamps = list(data_file_pd['date'])
```

```python
# Convert the timestamps to a date-hour format
dates = []
for ts in timestamps:
    conv = str(datetime.utcfromtimestamp(np.int32(ts)).strftime('%m/%d/%Y-%H:%M:%S'))
#    print(conv)
    dates.append(conv)
```

```python
# View the converted dates
dates
```

```python
# Create a dataf frame with the converted dates
dates_df = pd.DataFrame(dates, columns=["datetime"])
# View the results
dates_df.head()
```

```python
# Join the converted dates with the original data frame
final_df = pd.concat([data_file_pd,dates_df], axis=1, sort="False")
# View the results
final_df.head()
```

```python
# Export the city data into a .csv.
final_df.to_csv(output_data_file, index=False, header=True)
```
